**Ellipse plot**

$ \frac{x^2}{a^2} + \frac{y^2}{b^2} = 1 $

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = sp.symbols('x y')
#parameters of ellipse
a = 2
b = 1

#equation of ellipse
f = (x**2)/(a**2) + (y**2)/(b**2) -1 

#solve for y two halfs of the ellipse
y_upper = sp.solve(f,y)[0]
y_lower = sp.solve(f,y)[1]

#convert symbolic functions to numerical functions
y_upper_func = sp.lambdify(x,y_upper,'numpy')
y_lower_func = sp.lambdify(x,y_lower,'numpy')

x_vals = np.linspace(-a,a,500)

y1 = y_upper_func(x_vals)
y2 = y_lower_func(x_vals)

plt.plot(x_vals, y1, label='Upper half')
plt.plot(x_vals, y2, label='Lower half')
plt.gca().set_aspect('equal')
plt.title("Ellipse: x²/4 + y² = 1")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()

```
